```python
import pandas as pd

def convert_dtype(x):
    if not x:
        return ''
    try:
        return str(x)   
    except:        
        return ''

df = pd.read_csv('download.csv',converters={'Sustainable Development Goals': convert_dtype})
df.columns = df.columns.str.replace(' ', '')
df.columns = df.columns.str.replace('/', '')
df.index.rename('Key_Col', inplace=True)

df.head()
```

```python
# df.to_json('download.json')
df.to_csv('esg_clean.csv')
```
